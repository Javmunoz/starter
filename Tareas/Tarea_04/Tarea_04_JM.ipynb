```python
import numpy as np
import matplotlib.pyplot as plt
etiquetas = ["lunes", "martes", "miércoles", "jueves", "viernes", "sábado", "domingo"]
valores = [1.5, 1, 0, 0, 5, 0, 0.5]
valores_2 = [0, 0, 2, 2, 0, 5, 0]

co = np.arange(len(valores))
an = 0.30

fig, ax = plt.subplots()
ax.bar(co - an/2, valores, an, label="Guitarra", color="blue")
ax.bar(co + an/2, valores_2, an, label="Bajo", color="green")

ax.set_title("Horas dedicadas al bajo y la guitarra en una semana")
ax.set_ylabel("Horas")


ax.set_xticks (co)
ax.set_xticklabels(etiquetas)
plt.legend()

## Quiero aclarar que realicé una modificación de los datos de la entrega 3
#Utilicé la librería de Matplot porque fue la que me enseñaron a usar y me pareció muy intuitiva
#La historia que intento contar es la de una persona que gusta de tocar guitarra y bajo
#Con la información del gráfico se podrían sacar conclsiones como que está en una banda, estudia música, o le dedica su vida a la música, debido a que su practica es diaria
#Los días que más le dedicó a practicar fueron viernes y sábado
```
